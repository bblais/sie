```python
from sie.mcmc import StatsModel
```

```python
x=randn(100)*5+1.2
model=StatsModel()
model.add_data(x=x)
model.add("m ~ Normal(0,10)")
model.add("x ~ Normal(m,1)")
model.initialize()
```

```python
model.run_mcmc(800,repeat=2)
model.plot_chains()
```

```python
print(model.function_str)
```

```python
model.plot_distributions()
```

```python
xy1="""
X	Y
10	8.04
8	6.95
13	7.58
9	8.81
11	8.33
14	9.96
6	7.24
4	4.26
12	10.84
7	4.82
5	5.68
"""
xy=xy1
x,y=array([_.split() for _ in xy.strip().split('\n')[1:]],dtype=float).T
x,y
```

```python
model=StatsModel()
model.add_data(x=x,y=y)
model.add("m ~ Normal(0,10)")
model.add("b ~ Normal(0,10)")
model.add(" ~ Jeffreys()")
model.add("y ~ Normal(m*x+b,)")
model.initialize()
model
```

```python
print(model.function_str)
```

```python
model.run_mcmc(800,repeat=3)
model.plot_chains()
```
